### Bipartite Graph

```python
import matplotlib.pyplot as plt
import networkx as nx

color = ['gold', 'pink', 'tomato', 'darkseagreen', 'red']
fo = nx.Graph()
fo.add_nodes_from(['Friend1', 'Friend2', 'Friend3'], bipartite = 0)
fo.add_nodes_from(['Friend4', 'Friend5'], bipartite = 1)
fo.add_edges_from([('Friend4', 'Friend1'), ('Friend4', 'Friend2'), ('Friend4', 'Friend3'), ('Friend5', 'Friend1'), ('Friend5', 'Friend2'), ('Friend5', 'Friend3')])

top = nx.bipartite.sets(fo)[0]
pos = nx.bipartite_layout(fo, top)

nx.draw(fo, with_labels = True, pos=pos, node_color = color)
```

A bipartite graph is a type of graph whose vertices can be divided into two disjoint sets such that every edge connects a vertex from one set to a vertex in the other set. There are no edges connecting vertices within the same set. My example of a Bipartite graph comes from a true experience of still having lots of common friends with someone I cut off a year ago. It shows the connection between the friends(Friend3, Friend1, Friend2) and the lack of connection from each other(Friend4 and Friend5).

### Degree of a Vertex

```python
import matplotlib.pyplot as plt
import networkx as nx
G = nx.Graph()

G.add_nodes_from([1, 2, 3, 4, 5, 6, 7])
edgelist = [(1, 1), (3, 3), (1, 2), (4, 3), (2, 4), (2, 5), (5, 6), (5, 4)]
G.add_edges_from(edgelist)

pos = nx.planar_layout(G)
nx.draw(G, pos=pos,with_labels=True)
plt.show()
```

The graph has:
- Vertex 1: Degree of 2 (1 edge to itself and 1 edge to vertex 2)
- Vertex 2: Degree of 3 (1 edge to vertex 1, 1 edge to vertex 4, and 1 edge to vertex 5)
- Vertex 3: Degree of 2 (1 edge to itself and 1 edge to vertex 4)
- Vertex 4: Degree of 3 (1 edge to vertex 2, 1 edge to vertex 3, and 1 edge to vertex 5)
- Vertex 5: Degree of 3 (1 edge to vertex 2, 1 edge to vertex 4, and 1 edge to vertex 6)
- Vertex 6: Degree of 1 (1 edge to vertex 5)
- Vertex 7: Degree of 0

The degree of a vertex in a graph is the number of edges incident to that vertex, i.e., the number of edges connected to it. In a directed graph, the degree of a vertex is split into two types: the in-degree, which is the number of edges pointing towards the vertex, and the out-degree, which is the number of edges pointing away from the vertex.

## Euler's Circuit

```python
import matplotlib.pyplot as plt
import networkx as nx
seed = 1
n = [1, 2, 3, 4, 5]
e = [(1, 2), (2, 3), (3, 1),(1, 4), (4, 5), (5, 1)]

G = nx.DiGraph()
G.add_nodes_from(n)
G.add_edges_from(e)

pos = nx.spring_layout(G)
nx.draw(G, with_labels = True, pos=pos, node_color = 'grey')
```

An Eulerian circuit or Euler's Circuit, named after the Swiss mathematician Leonhard Euler, is a circuit that visits every edge of a graph exactly once and returns to its starting point. For a graph to have an Eulerian circuit, it must meet two conditions: it must be a connected graph and every vertex (node) in the graph must have an even degree, meaning the number of edges incident to each vertex must be even. In the example, the Euler Circuit goes 1 -> 2 -> 3 -> 1 -> 4 -> 5 -> 1.

## Euler's Trail

```python
import matplotlib.pyplot as plt
import networkx as nx
seed = 1
n = [1, 2, 3, 4, 5]
e = [(1, 2), (2, 3), (3, 4),(4, 2), (2, 5)]

G = nx.DiGraph()
G.add_nodes_from(n)
G.add_edges_from(e)

pos = nx.spring_layout(G)
nx.draw(G, with_labels = True, pos=pos, node_color = 'grey')
```

An Eulerian trail in a graph is a trail (a sequence of edges) that visits every edge exactly once. If the trail ends at the starting vertex, it is called an Eulerian circuit. An Eulerian circuit exists in a graph if and only if the graph is connected and every vertex has an even degree (the number of edges incident to the vertex). In the example, the Euler's Trail goes 1 -> 2 -> 3 -> 4 -> 2 -> 5.

## The travelling Salesman Problem

```python
import networkx as nx
import matplotlib.pyplot as plt
seed = 6

G = nx.Graph()
G.add_nodes_from(["A","B","C","D","E","F"])
edgelist = [("A", "B", 47), 
            ("A", "C", 25),
            ("A", "D", 12),  
            ("A", "E", 20),
            ("A", "F", 58),
            ("B", "C", 14),
            ("C", "D", 8),
            ("D", "E", 16),
            ("E", "F", 10),
            ("F", "B", 32)]
G.add_weighted_edges_from(edgelist)

pos = nx.spectral_layout(G)
nx.draw(G, pos=pos, with_labels=True)

edge_labels = nx.get_edge_attributes(G, 'weight')
nx.draw_networkx_edge_labels(G, pos=pos, edge_labels=edge_labels, font_color='red', font_size=8)


plt.show()
```

Possible Routes:
- A -> D -> C -> B -> F -> E -> A = 96
- A -> E -> F -> B -> C -> D -> A = 96
- A -> C -> B -> F -> E -> D -> A = 109
- A -> D -> E -> F -> B -> C -> A = 109
- A -> B -> F -> E -> D -> C -> A = 138
- A -> C -> D -> E -> F -> B -> A = 138
- A -> E -> D -> C -> B -> F -> A = 148
- A -> F -> B -> C -> D -> E -> A = 148
- A -> B -> C -> D -> E -> F -> A = 154
- A -> F -> E -> D -> C -> B -> A = 154

The best route is A -> D -> C -> B -> F -> E -> A or A -> E -> F -> B -> C -> D -> A with a distance of 96. The traveling salesman problems finds the shortest Hamiltonian circuit that is possible given a graph. The routes shown above are all possible Hamiltonian circuit and we can see the Hamiltonian circuit with the shortest distance of 96 and longest distance of 153.
